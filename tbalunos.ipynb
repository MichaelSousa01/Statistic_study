```python
import pandas as pd
dados_nome_nivel=pd.read_excel('/content/drive/MyDrive/grafic/dispersão/niveldoaluno(1).xlsx')
```

```python
# Corrigindo nomes de colunas
dados_nome_nivel.columns = dados_nome_nivel.columns.str.strip().str.replace('\n', '', regex=False)

# Renomeando para facilitar
dados_nome_nivel.rename(columns={
    'Em uma escala de 0 a 10, como você avalia o seu nível de conhecimento em Python?': 'Python',
    'Em uma escala de 0 a 10, como você avalia o seu nível de conhecimento em Excel?': 'Excel'
}, inplace=True)

# Convertendo para numérico (caso estejam como texto)
dados_nome_nivel['Python'] = pd.to_numeric(dados_nome_nivel['Python'], errors='coerce')
dados_nome_nivel['Excel'] = pd.to_numeric(dados_nome_nivel['Excel'], errors='coerce')
```

```python

dados_nome_nivel.head()
```

```python

colunapy=dados_nome_nivel['Python']
mediapy=colunapy.mean()
modapy=colunapy.mode()
medianapy=colunapy.median()
amplitudepy=colunapy.max()-colunapy.min()
variancia_populacaopy=colunapy.var(ddof=0)# da população
variancia_amostrapy=colunapy.var() #da amostra (padrão n-1)
desvio_padraopy=colunapy.std()
```

```python
print(f'Valor médio do conhecimento em Python dos alunos:         {mediapy:.2f}')
print(f'Valor da moda do conhecimento em Python dos alunos:       {modapy.iloc[0]}')
print(f'Valor da mediana do conhecimento em Python dos alunos:    {medianapy:.2f}')
print(f'Valor da amplitude do conhecimento em Python dos alunos:  {amplitudepy}')
print(f'Variância populacional do conhecimento em Python:         {variancia_populacaopy:.2f}')
print(f'Variância amostral do conhecimento em Python:             {variancia_amostrapy:.2f}')
print(f'Desvio padrão do conhecimento em Python:                  {desvio_padraopy:.2f}')
```

```python
frequencia_absoluta=colunapy.value_counts() #só com o .value_counts() tenho a fi em ordem decrescente de frequnacias e não em rol(ou seja em ordem crescente de valores)
print(frequencia_absoluta)
```

```python
frequencia_absoluta=colunapy.value_counts().sort_index() #com o values_cont() e sort_index() a frequencia vem em ordem crescente dos valores, como em rol, o que é bom
#quando queremos trabalhar com a ordem da frequencia de distribuição.
print(frequencia_absoluta)
```

```python
frequencia_relativa=(fequencia_absoluta/fequencia_absoluta.sum())*100
print(frequencia_relativa.round(2).astype(str)+'%') #aqui a frequencia_relativa foi convertida em string
```

# Comparando os dados por meio de gráficos.

###Gráfico de barras - média por tipo de conhecimento

```python
# Puxar as médias das colunas
medias_conhecimentos = dados_nome_nivel[['Excel', 'Python']].mean()

# Gráfico de barras
plt.figure(figsize=(6, 5))
medias_conhecimentos.plot(kind='bar', color=['green', 'orange'])
plt.title('Média do Nível de Conhecimento em Excel e Python')
plt.ylabel('Média')
plt.ylim(0, 10)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

###Gráfico de barras agrupadas - barra por alunos

```python

# Gráfico de barras lado a lado
plt.figure(figsize=(12, 6))
# construindo as posições do eixo x
x = np.arange(len(dados_nome_nivel))
largura = 0.4

plt.bar(x - largura/2, dados_nome_nivel['Excel'], width=largura, label='Excel', color='green')
plt.bar(x + largura/2, dados_nome_nivel['Python'], width=largura, label='Python', color='orange')
plt.xticks(x, dados_nome_nivel['Nome Completo'], rotation=90)
plt.ylabel('Nível de Conhecimento')
plt.title('Nível de Conhecimento em Excel e Python por Aluno')
plt.legend()
plt.tight_layout()
plt.show()
```

Gráfico de  Boxplot - distribuição dos níveis

```python
# Grafico de Boxplot
plt.figure(figsize=(8, 6))
dados_nome_nivel[['Excel', 'Python']].boxplot()
plt.title('Distribuição do Nível de Conhecimento')
plt.ylabel('Notas')
plt.grid(True)
plt.tight_layout()
plt.show()
```
